## EDA on Zomato dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("zomato.csv",encoding='latin-1')
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
[features for features in df.columns if df[features].isnull().sum()> 0]
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
df_country = pd.read_excel("Country-Code.xlsx")
df_country.head()
```

```python
df.columns
```

```python
final_df = pd.merge(df, df_country, on='Country Code',how='left')
```

```python
final_df.head(2)
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_val = final_df.Country.value_counts().values
```

```python
plt.pie(country_val[:3],labels= country_names[:3],autopct='%1.2f%%')
```

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x='Aggregate rating', y= "Rating Count", data = ratings)
```

```python
sns.barplot(x='Aggregate rating', y= "Rating Count", data = ratings,hue="Rating color",palette=['pink','red','orange','yellow','green','green'])
```

```python
##count plot
sns.countplot(x="Rating color",data=ratings,palette=['pink','red','orange','yellow','green','green'])
```
